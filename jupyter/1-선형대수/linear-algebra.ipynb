## 전에 학습한 보안관, 강도 문제를 Python으로 좌표에 Plot 그리기

이 문제는 이미 해를 알고 있는데, 그렇지 않다면 x, y축, t값을 구하는데 다양한 값을 대입해 봐야 한다.

이 문제처럼 답을 아는 상황이 아니면 그래프가 일치하는지 값과 극한값을 바꾸고, 교차점 위치를 조정해가며 찾아야 한다.

하지만 이미 보안관은 30분만에 75km 지점에서 강도를 잡는걸 알고있고 속도까지 알고 있으니 Plot을 그리는 실습을 해 본다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

시작, 끝 좌표를 0~40으로 설정, 시작과 끝점 사이의 좌표값은 1000개로 설정

```python
time = np.linspace(0, 40, 1000) # start, finish, n points
```

강도가 이동한 거리 = $d = 2.5t$


```python
d_r = 2.5 * time
```

보안관이 이동한 거리 = $d = 3(t - 5)$

```python
d_s = 3 * (time - 5)
```

강도와 보안관의 거리 함수를 가져왔으니 matplotlib Plot을 그려봅니다. (파이썬에서 가장 많이 사용되는 데이터 시각화 패키지)

```python
fig, ax = plt.subplots()
plt.title("Linear algebra - A Bang Robber Caught")
plt.xlabel("Time (in minutes)") # x축
plt.ylabel("distance (in km)") # y축

ax.set_xlim([0, 40]) # x축 좌표 시작, 끝점 설정
ax.set_ylim([0, 100]) # y축 좌표 시작, 끝점 설정

# 그래프 직선 색상 설정
ax.plot(time, d_r, c = 'green')
ax.plot(time, d_s, c='brown')

# 교차점 색 설정 및 선 스타일 --로 설정
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```
